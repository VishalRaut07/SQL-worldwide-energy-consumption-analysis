*The dataset does not explicitly mention measurement units. Since it is sourced from EIA (Environmental Impact Assessment (EIA) ), I treated the values as standardized metrics and focused on relative comparisons, trends, ratios, and per-capita calculations rather than absolute unit values. Wherever needed, I clearly stated assumptions.*

| Dataset            | Unit                  |
| ------------------ | --------------------------------- |
| *Population*        | Total population 
|                      (count - million,billion)          |
|------------------ | ------------------ | ------------------ |
| *GDP*                | GDP PPP (relative values) |
|                     currency units (e.g., US Dollars, Indian Rupees, Euros) |
|------------------ | ------------------ | ------------------ |
| *Emission*           | Total emissions (comparative) |
|                        Total emissions are generally measured in Metric Tons (t),  |     |                         Million  Metric Tons (MtCO2 or MtCO2eq), |
|                         or Gigatons (GtCO2) per year. |
|------------------ | ------------------ | ------------------ |
| *Consumption*        | Energy consumption (standardized) |
|                       meta watt,gigawatt,Terawatt-hours (TWh) per annum |
|------------------ | ------------------ | ------------------ |
| *Production*         | Energy production (standardized)|
|                        kilowatt-hours,megawatts (MW),gigawatts (GW)|
|------------------ | ------------------ | ------------------ |
| *Per Capita Metrics* | Calculated per person   |  
|represent averages per person, calculated by dividing total metrics    |                  | (e.g., GDP, income, consumption) by the total population|
